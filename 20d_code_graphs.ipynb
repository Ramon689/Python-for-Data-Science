```python
#intialize enviornment settings and packages

import io
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



#prevent dataframe truncation by adjusting pandas display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

```

```python
#race is limiting variable for districts that could be included in the final analysis
#create list of distircts that will be used to subset each of the other datasets

race_df = pd.read_csv('cc_race_20d.csv')

districts= race_df.District_Key.unique()

#list of districts used to subset other datasets
d_list = districts.tolist()

#print(d_list)
```

```python
#verify how many districts in the complete case of Race dataset are present
#in the Gender dataset


gender_df = pd.read_csv('cc_gender_8y.csv')

districts_02 = gender_df.District_Key.unique()

d_list_02 = districts_02.tolist()

#print(d_list_02)


common_CC_districts = []

for i in range(len(d_list_02)):
  if d_list_02[i] in d_list:
    #print(d_list_02[i])
    common_CC_districts.append(d_list_02[i])

print(len(common_CC_districts)) #20

#all districts were confirmed in both complete case datasets
```

```python
#subset of gender data set that has district values consistant with race dataset

df_g_20d = gender_df[gender_df.District_Key.isin(d_list)]

len(df_g_20d.District_Key.unique())

#export
df_g_20d.to_csv('cc_gender_20d.csv')
```

```python
#########################################
#compare race sub populations

#########################################

df_race = pd.read_csv('/content/cc_race_20d_U.csv')

#recode populations to White / Hispanic / Black

print(df_race.Population.value_counts())
#WHITE                        160
#BLACK OR AFRICAN AMERICAN     65
#BLACK                         40
#HISPANIC                      40
#HISPANIC OF ANY RACE          15


new_pop = []
for i in range(len(df_race)):
  if df_race.iloc[i,6] == 'BLACK OR AFRICAN AMERICAN':
    new_pop.append('BLACK')
  elif df_race.iloc[i,6] == 'BLACK':
    new_pop.append('BLACK')
  elif df_race.iloc[i,6] == 'HISPANIC OF ANY RACE':
    new_pop.append('HISPANIC')
  elif df_race.iloc[i,6] == 'HISPANIC':
    new_pop.append('HISPANIC')
  else:
    new_pop.append('WHITE')

df_race['New_pop'] = new_pop
df_race.head(10)

print(df_race.New_pop.value_counts())
#WHITE       160
#BLACK       105
#HISPANIC     55
```

```python
data = df_race

plt.figure(figsize=(15,8))
ax = sns.boxplot(x = data['Year_End'],
            y = data['Total_Tested'],
            hue = data['New_pop'])


ax.axes.set_title("Total Tested per Year grouped by Race",fontsize=30)
ax.set_xlabel("Year",fontsize=20)
ax.set_ylabel("Students",fontsize=20)

ax.legend( loc = 2, bbox_to_anchor = (1,1))
#display barplot
plt.show()

```
