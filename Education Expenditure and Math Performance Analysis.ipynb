```python
from ast import Break
df_schools_00.sort_values(by = ['District_Key', 'Year_End'], ascending = [True, True], na_position = 'first')
df_scores_00.sort_values(by = ['District_Key', 'Year_End'], ascending = [True, True], na_position = 'first')



Year_End = df_schools_00.Year_End.unique()
District_Key = df_schools_00.District_Key.unique()

results_df = pd.DataFrame()

#for each year_end query and individual district_key
for y in Year_End:
  #subset of df for specific year
  df_y_temp_score = df_scores_00[df_scores_00['Year_End']== y]
  df_y_temp_school = df_schools_00[df_schools_00['Year_End']== y]



  for d in District_Key:
    #subset of df for specific year and district key
    df_y_d_temp_score = df_y_temp_score[df_y_temp_score['District_Key'] == d]
    df_y_d_temp_school = df_y_temp_school[df_y_temp_school['District_Key'] == d]

    #print(df_y_d_temp_score)
    #print(df_y_d_temp_score)



    #iterate though length of df_school saving the values of one row to all
    #associated rows from df_score
    for i in range(len(df_y_d_temp_school)):
      temp_cross_df = df_y_d_temp_score
      #add column values from df_school
      temp_cross_df['Agency_Name'] = df_y_d_temp_school.iloc[i,2]#'Agency_Name'
      temp_cross_df['County'] = df_y_d_temp_school.iloc[i,3]#'County'
      temp_cross_df['City'] = df_y_d_temp_school.iloc[i,4]#'City'
      temp_cross_df['Zip'] = df_y_d_temp_school.iloc[i,5]#'Zip'
      temp_cross_df['Latitude'] = df_y_d_temp_school.iloc[i,6]#'Latitude'
      temp_cross_df['Longitude'] = df_y_d_temp_school.iloc[i,7]#'Longitude'
      temp_cross_df['Avg_Tot_Cur_Exp'] = df_y_d_temp_school.iloc[i,8]#'Avg_Tot_Cur_Exp'
      temp_cross_df['Avg_Tot_Exp'] = df_y_d_temp_school.iloc[i,9]#'Avg_Tot_Exp'

      #temp cross df is then appended together to create results df
      #print(temp_cross_df.columns)
      #print(temp_cross_df.info())

      results_df = results_df.append(temp_cross_df)
    #print(len(results_df))





len(results_df)


```
